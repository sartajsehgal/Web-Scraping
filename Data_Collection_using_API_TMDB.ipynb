```python
import requests
api_key = 'b191b9a855ffe40e9275277e54870312'
```

```python
def get_data(api_key,page_no):
    url='https://api.themoviedb.org/3/discover/movie?api_key='+str(api_key)+'&language=en-US&sort_by=popularity.desc&primary_release_year=2000&page='+str(page_no)+'&with_genres=Comedy&include_all_movies=true'
    data=requests.get(url)
    data_array=data.json()['results']
    result=[]
    for i in data_array:
        result.append(str(i['id'])+','+i['original_title'])
    return result
```

```python
import time
datas=[]
page_no=1
while len(datas)<300:
    time.sleep(10)
    res=get_data(api_key,page_no)
    for d in res:
        datas.append(d)
    page_no+=1
datas
```

```python
df=pd.DataFrame(datas,columns=['Movie_id, Movie_Name'])
df
```

```python
df.to_csv('movie_id_name.csv')
```

```python
similar_movies=[]
cnt=0
for d in datas:
    cnt+=1
    movie_id=d.split(',')[0]
    url_mov='https://api.themoviedb.org/3/movie/'+str(movie_id)+'/similar?api_key=b191b9a855ffe40e9275277e54870312&language=en-US&page=1'
    data_sm=requests.get(url_mov)
    data_sm_arr=data_sm.json()['results']
    flag=0
    for entry in data_sm_arr:
        if flag<5:
            if entry['id']<int(movie_id):
                val=str(entry['id'])+','+str(movie_id)
            else:
                val=str(movie_id)+','+str(entry['id'])
            flag+=1
            if val not in similar_movies:
                similar_movies.append(val)
        else:
            break
    if cnt%39==0:
        time.sleep(10)
similar_movies
```

```python
df1=pd.DataFrame(similar_movies,columns=['Movie_id, Similar_Movie_id'])
df1
```

```python
df1.to_csv('movie_id_sim_movie_id.csv')
```
